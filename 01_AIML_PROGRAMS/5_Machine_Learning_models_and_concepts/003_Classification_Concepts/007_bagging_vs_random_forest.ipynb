```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
X,y = make_classification(n_features=5, n_redundant=0, n_informative=5,n_clusters_per_class=1)
```

```python
df = pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target'] = y
print(df.shape)
df.head()
```

```python
bag = BaggingClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(bag.estimators_[0])
plt.show()
#as u see column sampling is done on tree level,,,ie;our baggingclassifier will sampled 'col1'(x[0]) and 'col2'(x[1]) for 1st model(bag.estimators_[0]) in base models
#thats why complete DT is constructed with only with these 2 features(col1 and col2)
```

```python
rf = RandomForestClassifier(max_features=2)
```

```python
rf.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(rf.estimators_[4])
plt.show()
#DT constructed by 5th model/estimator in base models  (rf.estimators_[4]),,,u can check for 1st model also (rf.estimators_[0])
#here complete DT is constructed using any column ie;sampled during column sampling
#column sampling is done at each node  ,, thats why at each node diff diff column/feature is slecetd
#which introduce more randomnesss in our model,,,,,thats why randomForest better than Bagging with base_estimator=DecisiontreeClassifier()...
```
