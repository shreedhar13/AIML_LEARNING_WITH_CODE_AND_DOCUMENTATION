# augment the dog.jpg

```python
import tensorflow
from tensorflow import keras
from keras.preprocessing import image  #behave like openCV
from keras.preprocessing.image import ImageDataGenerator
#keras-> image will do same work as cv2....not exactly same ,,but some features are there from cv2
```

```python
img=image.load_img(r"D:\DEEP_LEARNING\C_N_N\4_cats_vs_dog_classification\dog.jpg",target_size=(200,200))#in openCV imread() used ,,here load_img used
#and convert that image resolution to 200x200 at the time of loadings
```

```python
img  #if u use matplotlib then-->plt.imshow(img),,bcz opencv imread will read data in numpy array
#but this will load the image,,not load array
```

```python
datagen = ImageDataGenerator(
            rotation_range=30,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True,
            vertical_flip=True,
            width_shift_range=0.2,
            height_shift_range=0.2
)
#vertical_flip=True, -->u can remove it,,bcz inverted images are also created ,so if u dont want
#see parameter and try them
```

```python
img=image.img_to_array(img)#convert image to numpy array(3d array bcz colurfull image is theres)
```

```python
img.shape
```

```python
input_batch=img.reshape(1,200,200,3)#as u see in cat_vs_dog_classification  we are giving i/p in batch,,bcz of ram shortage,,
#so here only one image is present,,thus make batch has 1 image data and pass it to i/p
```

```python
i=0

for output in datagen.flow(input_batch,batch_size=1,save_to_dir='augmented_images'):
    #we use flow-->when we have only 1 image(numpy array representaion of image) to process
    #flow_from_directory -->when u have multiple image (numpy array representaion of image) in directory to process then use this
    i = i+1
    if i==10:
        break
#,batch_size=1,--->each created(augmented) image is consider as 1 batch
#if batch_size=2 -->  2 augmeneted image are considered as 1 batch and it is passed at a time to model as i/p
#if batch_size=3 -->  3 augmeneted image are considered as 1 batch and it is passed at a time to model as i/p
```

```python
#see this augmented_images folder 10 augmented images are formed and each has its own batch ie;10 batches and each has 1 image data
#therefore as data is augmented/creted ,that one is passed as i/p to model bcz batch_size=1
#if batch_size-2,,then it waits untill 2 augmented images to create,,when created then pass both at a time to model as i/p,,thus 5 batches atre formed and each batch with 2 image data
```
